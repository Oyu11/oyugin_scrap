```python
import requests
import urllib.request
from bs4 import BeautifulSoup
import pandas as pd
```

```python
class Product:
    def __init__(self, une_, motor_, hurdnii_hairtsag_, hurd_, turul_, ungu_, uild_on_, orj_on_, hudulguur_, dotor_ungu_, leasing_, hutlugch_, kilo_ , nuhtsul_,  haalga_ ):
        self.une= une_
        self.motor = motor_
        self.hurdnii_hairtsag = hurdnii_hairtsag_
        self.hurd = hurd_
        self.turul = turul_
        self.ungu = ungu_
        self.uild_on = uild_on_
        self.orj_on = orj_on_
        self.hudulguur = hudulguur_
        self.dotor_ungu = dotor_ungu_
        self.leasing = leasing_
        self.hutlugch= hutlugch_
        self.kilo= kilo_
        self.nuhtsul = nuhtsul_
        self.haalga = haalga_
```

```python
baseurl = 'https://www.unegui.mn/avto-mashin/-avtomashin-zarna/mercedes-benz/'
product_list = []
for i in range(1, 8):
    url = baseurl
    #print(url)
    response = requests.get(url)
    if response.status_code != 200:
        print(response.status_code)
        print('error ', url)
        continue
    #print(response.text)
    soup = BeautifulSoup(response.text, "html.parser")
    li_list = soup.find_all("div", {"class": "swiper-wrapper"})
    for li in li_list:
        a = li.find('a')
        product_url = "https://unegui.mn" + a['href']
        print(product_url)
        product_list.append(product_url)
```

```python
import re
def extract_number(text):
    # Use regular expression to find the number
    match = re.search(r'\d+\.?\d*', text)
    if match:
        return match.group()
    else:
        return None
```

```python
def find_Info(div_list, header):
  ref = None
  for li in div_list:
    txt = li.text.strip()
    if txt.startswith(header):
      ref = txt[len(header):].strip()
      break
  return ref
```

```python

df = pd.DataFrame(product_data)
df.to_csv('car1.tsv', sep="\t", index=False)
```
